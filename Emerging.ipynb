# Data Loading

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('Iowa_Liquor_Sales.csv')
```

# Basic EDA

```python
#To check the shape of the dataset
data.shape
```

```python
#To check the data-types of each coloumn
data.dtypes
```

```python
#To check null values 
data.isnull().sum()
```

```python
#Dropping null values in the coloumn County
data.dropna(subset=['County'])
```

```python
#Checking shape of the Dataset
data.shape
```

```python
#Extracting only the data of December month of all years and saving it in data_dec
data['Date'] = pd.to_datetime(data['Date'])
data['Month']=  data['Date'].dt.month
data_dec = data.loc[data['Month'] == 12]
data_dec
```

```python
#Creating an object and storing only required columns
dfObj = pd.DataFrame(data_dec, columns=['Item Number','Item Description','County','Store Number','Pack','Bottles Sold'])
# Select duplicate rows except first occurrence based on County and Store Number columns
df = dfObj[dfObj.duplicated('County', keep = 'first')]
df = dfObj[dfObj.duplicated('Store Number', keep = 'first')]
df.shape
```

```python
df
```

```python
#Creating a coloumn which has the Stock left numbers of each store and each item
sum_column = df["Pack"] - df["Bottles Sold"]
df["Stock Left"] = sum_column
df
```

```python
#To check null values 
df.isnull().sum()
```

```python
df.dropna(axis = 0,subset=['County'])
```

```python
df.isnull().sum()
```

```python
sns.countplot(df['Stock Left'])
```

```python
sns.countplot(df['Bottles Sold'])
```

```python
#Performing Correlarion matrix to see what all features affect on the outcome
fig , ax = plt.subplots(figsize = (10,7))
sns.heatmap(df.corr() , annot = True , ax = ax)
plt.show()
```
